# Robust Regression

Linear regression is heavily influenced by outliers.

Outlier demo: http://digitalfirst.bfwpub.com/stats_applet/stats_applet_5_correg.html

```python
df.head()
```

# RANdom SAmple Consensuc (RANSAC) Algorithm

link = https://scikit-learn.org/stable/modules/linear_model.html#ransac-regression

Each iteration performs the following steps:
1. Select min_samples random samples from the original data and check whether the set of data is valid (see is_data_valid).
2. Fit a model to the random subset (base_estimator.fit) and check whether the estimated model is valid (see is_model_valid).
3. Classify all data as inliers or outliers by calculating the residuals to the estimated model (base_estimator.predict(X) - y) - all data samples with absolute residuals smaller than the residual_threshold are considered as inliers.
4. Save fitted model as best model if number of inlier samples is maximal. In case the current estimated model has the same number of inliers, it is only considered as the best model if it has better score.

```python
X = df['RM'].values.reshape(-1, 1)
y = df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3,10,1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,10))
plt.scatter(X[inlier_mask], y[inlier_mask],
           c='blue', marker='o', label='Inlier')
plt.scatter(X[outlier_mask], y[outlier_mask],
           c='brown', marker='s', label='Outlier')
plt.plot(line_X, line_y_ransac, color='red') # best fitted line
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median value of owner occupied homes in 1000s of dollars")
plt.legend(loc='upper left')
plt.show();
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

```python
X = df['LSTAT'].values.reshape(-1, 1)
y = df['MEDV'].values
ransac = RANSACRegressor()
ransac.fit(X, y)
```

## Now we are going to evaluate the model using different methods

# Method 1: Residual Analysis

```python
plt.figure(figsize=(12,10))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Testing data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='k')
plt.xlim([-10, 50])
plt.show();
```

If your model would be able to explain all the variability of your data, then normally you would not have outliers. But in this instance we do have outliers, so there are some data that is not explained by our model. So if we look at the residuals there is a certain form of structure to the data, not linear, but some structure.

# Method 2: Mean Squared Error (MSE)

$$MSE=\frac{1}{n}\sum^n_{i=1}(y_i-\hat{y}_i)^2$$

* The average value of the Sums of Squared Error cost function  

* Useful for comparing different regression models 

* For tuning parameters via a grid search and cross-validation

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

Observation: the models performs better insample (the error is 36.5) than outsample (the erorr is 46.3), it's gone up.

# Method 3: Coefficient of Determination, $R^2$

$$R^2 = 1 - \frac{SSE}{SST}$$

SSE: Sum of squared errors

SST: Total sum of squares

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

With r2 score the higher the number the better it is, so we have a 0.57 for train and then the r2 score goes down for testing. r2 has a value between 0 and 1, 1 means that the model can perfectly explain the data, 0 means that the model can not explain anything about the data.

# What does a near perfect model look like?

To have a point of reference for those metrics and see how they would evaluate on a near perfect model, a model that can almost perfectly describe the data we are going to generate some artificial data and play with it.

```python
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(1000)
y = 3 * x + np.random.randn(1000)
plt.figure(figsize=(12,10))
plt.scatter(x, y)
plt.show();
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(X_train.reshape(-1,1), y_train)

y_train_pred = model.predict(X_train.reshape(-1,1))
y_test_pred = model.predict(X_test.reshape(-1,1))
```

# Method 1: Residual Analysis

```python
plt.figure(figsize=(12,10))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Testing data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-3, xmax=33, lw=2, color='k')
plt.xlim([-5, 35])
plt.ylim([-25, 15])
plt.show();
```

The variability of the data (on the residuals axis) is substantially reduced compared to the previous graphic for the housing data, now it is between [-3,3]. So a near perfect model looks like this concentrated/focused in a small place, with not so high variability and differences between the testing and training data.

# Method 2: MSE

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

# Method 3: Coefficient of Determination, $R^2$

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

This shows clearly that 98.67% of data in sample is explained by the model, and 98.62% of outside of sample data is explained by the model. Remember that higher values of r2 score are better. Compared with the values for the housing data those are way better.
